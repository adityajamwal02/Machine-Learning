```python
# Date on which house was issued (getting only year part) ---> Brute Loop 


year=[]
for i in range(len(data['Date House was Sold'])):
    k=data['Date House was Sold'][i].split()[-1]
    year.append(k)
    
data['year_sold']=year
data['year_sold'].head()
```

```python
# Date on which house was issued (getting only year part) using pandas.map()

def year(value):
    return value.split()[-1]
data['year_sold']=data['Date House was Sold'].map(year)
data['year_sold'].head()
```

```python
# Defining Luxury home based on any 2 conditions out of 3 to be true. --> Brute Force loop


def luxury_home(row):      # Function defining 
    count=0
    if row[0]=='Yes':
        count+=1
    if row[1] in ['Good', 'Excellent']:
        count+=1
    if row[2]>=8:
        count+=1
    if count>=2:
        return "yes"
    else:
        return "no"

data['luxury_home']=data[['Waterfront View', 'Condition of the House', 'Overall Grade']].apply(luxury_home, axis=1)
data['luxury_home'].unique()
```

# OUTLIER



```python
data['Sale_Price'].describe()
```

```python
# Plotting to find an outlier through visualization

plt.scatter(x=data['ID'],y=data['Sale Price'])
plt.show()
```

```python
# Interquartile Range (Q3-Q1) & Setting lower limit

q1=data['Sale_Price'].quantile(0.25)
q3=data['Sale_Price'].quantile(0.75)
iqr=q3-q1
lower_limit=q1-1.5*iqr
print(lower_limit)
```

```python
# Setting upper limit

upper_limit=q1+1.5*iqr
print(upper_limit)
```

```python
# Assigning min value to lower_limit

lower_limit=min(data['Sale Price'])
print(lower_limit)
```

```python
# Importing a library to ignore the outlier and display boxplot

import seaborn as sns
sns.boxplot(x=data['Sale Price'])
```

```python
# Treatment of an Outlier by imputing method

lower_limit=min(data['Sale Price'])
upper_limit=q3+1.5*iqr

def limit_inputer(value):
    if value>upper_limit:
        return upper_limit
    if value<lower_limit:
        return lower_limit
    else:
        return value

data['Sale Price']=data['Sale Price'].apply(limit_inputer)
```

```python
# Outlier as been treated, now check for it

data['Sale Price'].describe()
```

```python
# Treating a missing value using deletion (dropna fnc used)

data.dropna(inplace=True,axis=0,subset=['Sale Price'])
```

# ANOVA



```python
# Importing Module for ANOVA

from statsmodels.formula.api import ols
import statsmodels.api as sm


# Renaming the variables with no space using replace fn

data=data.rename(columns={'Sale Price':'Sale_Price','Condition of the House':'Condition_of_the_House','Ever Renovated':'Ever_Renovated','Waterfront View':'Waterfront_View'})

# Checking if the renaming was done correctly 
data.head(4)
```

```python
# Executing ANOVA for condition of the house 

mod=ols('Sale_Price ~ Condition_of_the_House', data = data).fit()
```

```python
# Creating one way ANOVA table using Stats library and then printing it over

Anova_Table=sm.stats.anova_lm(mod, typ=2)
print(Anova_Table)
```

```python
# Creating one way ANOVA table using Stats library and then printing it over (Waterfront View)

mod=ols('Sale_Price ~ Waterfront_View', data=data).fit()
sm.stats.anova_lm(mod,typ=2)
```

```python
# Creating one way ANOVA table using Stats library and then printing it over (Zipcode)

mod=ols('Sale_Price ~ Zipcode', data=data).fit()
sm.stats.anova_lm(mod,typ=2)
```

# Binning


```python
# Getting avg sale price across each zipcode

Zip_Table=data.groupby('Zipcode').agg({'Sale Price':'mean'}).sort_values('Sale Price', ascending=True)
```

```python
Zip_Table.head()
```

```python
# Grouping Zipcodes into 5 bins

Zip_Table['Zipcode_Group']=pd.cut(Zip_Table['Sale_Price'], bins=5,
                                 labels=['Zipcode_Group_0','Zipcode_Group_1','Zipcode_Group_2','Zipcode_Group_3','Zipcode_Group_4'], include_lowest= True)
```

```python
# Dropping Columns 

Zip_Table=Zip_Table.drop(columns='Sale Price')
```

```python
# Merging Tables using merge fnc. with left join over Zipcode Column

data=pd.merge(data,Zip_Table,left_on='Zipcode',how='left',right_index=True)
```

```python
# Dropping the Zipcode column since it is already included once in merge above

data=data.drop(columns="Zipcode")
```

```python
# Creating dummy variables for Zipcode group

data=pd.get_dummies(data,
                    columns='Zipcode_Group',
                    drop_first=True)
```
